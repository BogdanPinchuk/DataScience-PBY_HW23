# ✅ Рішення: 1 — Робота з HuggingFace (обов’язкове)

1. Встановіть бібліотеку `transformers`
2. Завантажте будь-яку модель через `pipeline` для   sentiment-analysis
3. Зробіть 3 різні запити:
   - позитивний текст
   - негативний текст
   - нейтральний текст

Результат:
- вивід результатів моделі
- короткий висновок (2–3 речення): чи правильно модель усе класифікує

```python
# Classifier

import torch
import apps.reporter as rpt
from pathlib import Path
from transformers import pipeline

# Input data
load_model = True
model_file_name = "resources/sentiment_model"
input_texts = [
    "Цей курс Data Science просто неперевершений, він надає корисні фундаментальні знання!",
    "Найгірші компанії — це ті, які з року в рік підіймають ціни, але не дають ніяких інновацій, тому їх крах неминучий.",
    "Напрямок Data Science прогнозовано розвивається з року в рік.",
    "Технічні науки - дуже цікава сфера знань.",
    "Через стрічки в соцмережах ми сильно деградуємо й марно витрачаємо свій час.",
]

# Solution
file_path = Path(model_file_name)

# try to load model
if load_model and file_path.exists():
    sentiment_model = pipeline(
        "text-classification",
        model=model_file_name,
        tokenizer=model_file_name
    )
else:
    sentiment_model = pipeline(
        "text-classification",
        model="tabularisai/multilingual-sentiment-analysis"
    )
    # save model
    sentiment_model.model.save_pretrained(model_file_name)
    sentiment_model.tokenizer.save_pretrained(model_file_name)

rp1 = rpt.Reporter("Запис", "Оцінка")
rp1.tolerance = 4

for text in input_texts:
    response = sentiment_model(text)[0]
    rp1.add_item(text, f"Label: {response["label"]}, Score: {rp1.format_value(response["score"])}")

# Print results
rp1.print_pd_report("Результати моделі")
```

Висновок. Для роботи з українською мовою була використана багатокласова мультимовна модель, яка чудово все покласифікувала. Але вона не достатньо хороша, і інколи без більшого контексту негативні результати бачить як позитивні.

# ✅ Рішення: 2 — Генерація тексту

Використайте модель для генерації тексту.

1. Створіть prompt (будь-який):
   - про AI
   - про навчання
   - про майбутнє технологій

2. Зміні параметри:
   - `temperature = 0.2`
   - `temperature = 0.8`
   - `temperature = 1.2`

Результат:
- 3 різні відповіді
- пояснення: як змінюється стиль тексту

```python
# Text Generation

import torch
import apps.reporter as rpt
from pathlib import Path
from transformers import pipeline, GenerationConfig

# Input data
load_model = True
model_file_name = "resources/text_generation_model"
temperatures = [0.2, 0.8, 1.2]
prompt = "У майбутньому генеративний штучний інтелект буде"

# Solution
file_path = Path(model_file_name)

# try to load model
if load_model and file_path.exists():
    generator = pipeline(
        "text-generation",
        model=model_file_name,
        tokenizer=model_file_name,
        clean_up_tokenization_spaces=False,
    )
else:
    generator = pipeline(
        "text-generation",
        model="benjamin/gpt2-wechsel-ukrainian",
        # model="malteos/gpt2-uk",
        clean_up_tokenization_spaces=False,
    )
    # save model
    generator.model.save_pretrained(model_file_name)
    generator.tokenizer.save_pretrained(model_file_name)

rp1 = rpt.Reporter("Коефіцієнт креативності", "Відповідь")

for temperature in temperatures:
    config = GenerationConfig(
        max_new_tokens=50,
        temperature=temperature,
        num_return_sequences=1,
        do_sample=True,  # випадковий вибір слів
        top_p=0.9,  # уникаємо малоймовірних слів
        no_repeat_ngram_size=3,  # уникаємо зациклювання
        pad_token_id=generator.tokenizer.eos_token_id,
    )
    response = generator(prompt, generation_config=config)[0]["generated_text"]
    rp1.add_item(str(temperature), response)

# Print results
rp1.print_pd_report("Результати моделі")
```

Висновок. Зі збільшенням коефіцієнта креативності модель робить менше повторень і генерує більш різноманітний текст, проте його зміст стає менш логічним та зв'язним.

# ✅ Рішення: 3 — Prompt Engineering

Створіть 3 варіанти prompt для однієї задачі:

Задача:
> Класифікація настрою тексту

Варіанти:
1. Zero-shot prompt
2. Few-shot prompt (з 2 прикладами)
3. Role prompt (наприклад: “You are an AI assistant...”)

Результат:
- 3 промпти
- результати моделі
- який варіант працює найкраще і чому

```python
# Prompt Engineering

import torch
import apps.reporter as rpt
from transformers import pipeline, GenerationConfig

# Input data
load_model = True
model_file_name = "resources/instruction_understanding_model"

prompts = {
    "Zero-shot": """
    Класифікуй настрій.
    Поверни ТІЛЬКИ одне слово: "Позитивний" або "Негативний".
    Текст для аналізу: Сервіс має баги.
    Настрій:
    """,
    "Few-shot": """
    Класифікуй настрій.
    Текст: "Цей метод хороший"
    Настрій: Позитивний
    Текст: Цей код поганий.
    Настрій: Негативний
    Текст для аналізу: Сервіс має баги.
    Настрій:
    """,
    "Role prompt": """
    Ви помічник зі штучним інтелектом.
    Поясни, що таке Data Science, простими словами для новачка:
    Відповідь:
    """
}

# Solution
file_path = Path(model_file_name)

# try to load model
if load_model and file_path.exists():
    generator = pipeline(
        "text-generation",
        model=model_file_name,
        tokenizer=model_file_name,
        clean_up_tokenization_spaces=False,
    )
else:
    generator = pipeline(
        "text-generation",
        model="Qwen/Qwen2.5-0.5B-Instruct",
        # model="Qwen/Qwen2.5-1.5B-Instruct",
        clean_up_tokenization_spaces=False,
    )
    # save model
    generator.model.save_pretrained(model_file_name)
    generator.tokenizer.save_pretrained(model_file_name)

rp1 = rpt.Reporter("Промпт для моделі", "Відповідь")

for prompt_name, prompt in prompts.items():
    if prompt_name == "Role prompt":
        max_new_tokens = 55
    else:
        max_new_tokens = 10

    config = GenerationConfig(
        max_new_tokens=max_new_tokens,
        temperature=0.3,
        do_sample=False,
        no_repeat_ngram_size=3,
        pad_token_id=generator.tokenizer.eos_token_id,
    )
    response = generator(prompt, generation_config=config)[0]["generated_text"][len(prompt):].strip()
    rp1.add_item(f"{prompt_name}\n{prompt}", response)

# Print results
rp1.print_pd_report("Результати моделі")
```

Висновок. Ця мультимовна модель показала себе доволі погано, лише в Role prompt вона спробувала зробити пояснення. Але причиною є те, що через обмеження памʼяті, обрана модель займає менше місця, а отже й має гірше навчання. У випадку вибору більш потужнішої моделі, результати були б набагато кращі, але при цьому й обчислювальних ресурсів треба буде більше.

# ✅ Рішення: 4 — Ембеддінги

1. Візьміть 4–6 текстів (наприклад, фрази про технології)
2. Побудуйте embeddings (можна через `sentence-transformers`)
3. Порахуйте cosine similarity
4. Знайдіть:
   - найбільш схожі пари текстів

Результат:
- матриця схожості
- 2–3 висновки

```python
# Embeddings and Sentence Matching

import torch
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer, util

# Input data
load_model = True
model_file_name = "resources/embedding_model"

sentences = {
    0: "Оптичні процесори використовують фотони замість електронів для надшвидкісних обчислень.",
    1: "Фотонні чипи та світлові технології дозволяють кардинально підвищити швидкість обробки даних.",
    2: "Сучасні ігрові рушії Unreal Engine дозволяють створювати реалістичну графіку.",
    3: "Розробка відеоігор на базі сучасних рушіїв потребує оптимізації графіки та ігрового процесу.",
    4: "Колонізація Марса стане головним технологічним проривом людства в цьому столітті.",
    5: "Міжпланетні польоти та будівництво баз на місяці змінять майбутнє цивілізації.",
}

# Solution
file_path = Path(model_file_name)

# try to load model
if load_model and file_path.exists():
    model = SentenceTransformer(model_file_name)
else:
    model = SentenceTransformer('sentence-transformers/LaBSE')
    # save model
    model.save(model_file_name)

rp1 = rpt.Reporter("ID речення", "Речення")
for idx, sentence in sentences.items():
    rp1.add_item(str(idx), sentence)

embeddings = model.encode(list(sentences.values()), convert_to_tensor=True)

# variant 1
similarity = util.cos_sim(embeddings, embeddings).cpu().numpy().round(3)
# # variant 2
# similarity = cosine_similarity(embeddings.cpu().numpy()).round(3)

df_sentence_corr = pd.DataFrame(similarity)
df_sentence_corr = df_sentence_corr.replace(1.0, 0.0)

rp2 = rpt.Reporter("Оригінальне речення", "Подібне речення")

skip_id = []
for col_name_id in df_sentence_corr.columns:
    if col_name_id not in skip_id:
        col_series = df_sentence_corr[col_name_id]
        similar_id = col_series.idxmax()
        skip_id.append(similar_id)
        rp2.add_item(str(sentences.get(int(col_name_id))), str(sentences.get(int(similar_id))))  # type: ignore

# Print results
rp1.print_pd_report("Речення для моделі")
display(df_sentence_corr.style.set_caption("Матриця подібності (модифікована)"))
rp2.print_pd_report("Подібні пари речень")
```

Висновок. Модель успішно зміст фраз і правильно обʼєднала попарно за темами оптичні процесори, розробка ігор, і космос. Косинусна дистанція доволі хороша між відмінними темами, хоча деяка схожість залишається, оскільки модель розпізнала спільний технічний контекст.
